#### 정규표현식 문법 리스트 -1
- [0-9] : 숫자를 찾음 => \d (*decimal)
- [^0-9] : 숫자가 아닌 값  ^은 아니다라는 뜻 (*문자, 특수문자, 공백 등) => \D
- [\t\n\r\f\v] : 탭과 같은 여백, 줄바꿈 등의 문자를 찾고자 할 때 => \s
- [^\t\n\r\f\v] : 탭과 같은 여백, 줄바꿈 등의 문자 외의 값을 찾고자 할 때 (*문자, 특수문자, 숫자 등) => \S
- [A-Za-z0-9] : 문자 혹은 숫자를 찾음 => \w
- [^A-Za-z0-9] : 문자 혹은 숫자 외의 값을 찾고자 할 때 => \W

#### 정규표현식 문법 리스트 -2
- Dot = . : 문자 1개를 의미
- ? : 앞에 있는 문자가 0번 또는 1번 표시되는 패턴 = 없어도 되고, 있따면 한번만 있어야 하는 패턴
- + : 앞에 있는 문자가 1번 또는 그 이상 반복되는 패턴
- * : 앞에 있는 문자가 0번 또는 그 이상 반복되는 패턴

#### 정규표현식 문법 리스트 -3
- {n} : 앞에 있는 문자가 n번 반복되는 패턴
- {m,n} : 앞에 있는 문자가 m번 반복되는 패턴부터 n번 반복되는 패턴까지

#### 정규표현식 문법 리스트 - 4
- [] : 대괄호 내부에 입력한 값을 "또는"의 형태로 찾아올 수 있는 패턴
- ex) [abc] : a 또는 b 또는 c 중 하나가 들어가 있는 패턴
- ex) [abcdefg]
- "-" : 패턴 내 모든 문자를 일일히 다 기입할 수 없기 때문에
- [a-zA-Z] : 소문자 알파벳 전체 또는 대문자 알파벳 전체

#### 정규표현식 문법 리스트 - 5
- 한글만 정규표현식의 패턴으로 정의 : [가-힣]

#### 정규표현식 문법 리스트 - 6
- match함수 : 문자열의 처음부터 끝까지 정규표현식과 매칭되어지는 문자만을 찾아서 리턴
- search함수 : 문자열 전체 중에서 정규표현식과 매칭되어지는 문자가 있으면 찾아서 리턴
- findall함수 : 정규표현식 패턴과 매칭되는 모든 문자열을 리스트 자료형태로 리턴
- split함수 : 정규표현식 패턴에서 구분자로 입력된 문자열을 기준으로 문자열을 분리

#### 정규표현식 미션
- 801210-1011323 라는 주민등록번호를 확인 후 가운데 -를 *로 교체하여 출력해주세요.

```python
import re

word ="801210 - 1011323"
pattern15 = re.compile("-")
subed = pattern15.sub("*", word)

print(subed)
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re #정규표현식의 방식을 쓰겠다.

res = urlopen("https://crawlingtest.netlify.app/")
soup = BeautifulSoup(res, "html.parser")

datas = soup.select("ul#dev_course_list > li.course")
for data in datas : 
    print(re.sub("\[[0-9]+\]", "",data.get_text()).strip()) #앞에 [패턴] 뒤에 []문자 # escape sequence => \n \t
    # print(data.get_text().strip())
```

```python
import openpyxl
import re

regex = re.compile(" [A-Za-z]+\.") #대문자..소문자...그걸 패턴으로 하고 온점이 붙을거다

work_book = openpyxl.load_workbook("train.xlsx")
work_sheet = work_book.active

for each_row in work_sheet.rows :
    print(each_row[3].value)
    print(regex.findall(each_row[3].value))
```
