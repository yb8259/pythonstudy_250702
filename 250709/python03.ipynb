```python
import requests #라이브러리 객체 가져와서
from bs4 import BeautifulSoup #라이브러리 객체 가져와서 beautifulsoup 빼옴

res = requests.get("https://v.daum.net/v/20170615203441266") 
soup = BeautifulSoup(res.content,"html.parser") # res.content 의 html을 파싱해줘!
mydata = soup.find("title") #나온값 : <title>잔금대출에도 DTI 규제 적용 검토</title>
print(mydata.get_text())
```
