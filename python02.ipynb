img를 찾아오고 리스트형태에 append하기, urlretrieve활용해서 url 수집후 이미지 저장하기

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service # 옵셥값 정의
from selenium.webdriver.common.by import By #~에 의한 명령
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from urllib.request import urlretrieve #url을 수집하다~ 그리고 저장할거야
import time
import pyautogui #Python에서 마우스, 키보드 자동 제어를 할 수 있게 해주는 강력한 라이브러리를 불러오는 코드입니다.

service = Service(ChromeDriverManager().install())
options = Options()
# options.add_argument("headless")
options.add_argument("window-size-1920x1080")
options.add_argument("--disable-gpu")
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36")
options.add_argument("lang=ko_KR")

driver = webdriver.Chrome(service=service, options=options)
driver.get("https://davelee-fun.github.io/")

elements = driver.find_elements(By.CSS_SELECTOR,"div.maxthumb img img-fluid") #img 만 써도 됨.
sources = list() #리스트를 만들기 위한 for문 밖 전역변수! 

for element in elements :
    sources.append(element.get_attribute("src")) #찾아온 속성 src 을 차곡차곡 추가할거야

for index, source in enumerate(sources) :
    urlretrieve(source, f"IMG/{str(index)}.{source.split(".")[-1]}")

time.sleep(3)
print("완료")
driver.quit()
```

#각각의 사이트가 만들어진 방식이 다르기 때문에 XPATH 방식으로 사용합니다.
#### HTML 크롤링을 위한 XPATH 기본 포맷 (*형태)
- //tagname[@Attribute="value"]
- / : 절대경로를 의미  ex. html/body/div/.../... -> 엄청 자주 쓰지 않아요!
- // : 문서 내에서 바로 검색을 하라는 의미 ex. //h1 -> 문서에서 바로 h1태그를 찾아와라)
- //*[@href] : 문서 내 href라는 속성을 가지고 있는 모든 태그를 다 찾아와라 (근데 거의 앵커밖에 없긴함ㅋ)
- ex. //a[@href="https://www.google.com"] : 문서 내 앵커태그를 찾아오는데, href 속성값으로 구글 닷컴을 찾아와라
- (//a)[3] : 문서 내 a태그를 찾아오는데 세번째 a만 찾아와라 (문서상에선 1번부터 start!!)
- (//a)[positon() < 3] : 문서 내 a태그를 찾아오는데, 앞에서부터 2개만 찾아와라
- (//a)[position() mod 2 = 0] : 문서 내 a태그를 찾아오는데, 짝수만 찾아오세요
- (//a)[position() mod 2 = 1] : 문서 내 a태그를 찾아오는데, 홀수만 찾아오세요
- (//table)[last()] : 문서 내 table 태그를 찾아오는데, 마지막번째 table만 찾아와라
- //div[@*] : 문서 내 div태그를 찾아오는데, 어떤 속성이든 상관없이 속성이 하나라도 있으면 찾아와라

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service # 옵셥값 정의
from selenium.webdriver.common.by import By #~에 의한 명령
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from urllib.request import urlretrieve #url을 수집하다~ 그리고 저장할거야
import time
import pyautogui #Python에서 마우스, 키보드 자동 제어를 할 수 있게 해주는 강력한 라이브러리를 불러오는 코드입니다.

service = Service(ChromeDriverManager().install())
options = Options()
# options.add_argument("headless")
options.add_argument("window-size-1920x1080")
options.add_argument("--disable-gpu")
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36")
options.add_argument("lang=ko_KR")

query = "일본 여행 사진"
driver = webdriver.Chrome(service=service, options=options)
driver.get(f"https://cafe.daum.net/subdued20club/ReHf/5387486?q={query}&re=1")

elements = driver.find_elements(By.CSS_SELECTOR,"div.figure-img") #img 만 써도 됨.
sources = list() #리스트를 만들기 위한 for문 밖 전역변수! 

for element in elements :
    sources.append(element.get_attribute("src")) #찾아온 속성 src 을 차곡차곡 추가할거야

for index, source in enumerate(sources) :
    urlretrieve(source, f"IMG/{str(index)}.{source.split(".")[-1]}")

time.sleep(3)
print("완료")
driver.quit()
```
