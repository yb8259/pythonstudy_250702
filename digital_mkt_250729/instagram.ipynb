```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import pickle #쿠키값 저장, 불러오는 친구
import time 

options = Options()
options.add_argument("--start-maximized")
options.add_argument("window-size=1920x1080")
options.add_argument("lang=ko_KR")

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

driver.get("https://www.instagram.com")

with open("instagram_cookies.pkl", "rb") as f : # 어떠한 값을 가져와서 rb(read binary)해줄건데 f라는 이름으로 쓰겠다..?
    cookies = pickle.load(f) # load()는 파일에서 객체를 역직렬화해서 메모리로 복원
for cookie in cookies : 
    driver.add_cookie(cookie)

driver.refresh()
time.sleep(5)

try : 
    not_now = driver.find_element(By.XPATH,"//button[text()='나중에하기']")
    not_now.click()
    time.sleep(2)
except :
    pass

keyword = "무신사"
driver.get(f"https://www.instagram.com/explore/tags/{keyword}")
time.sleep(5)

post_links = set() # 중복값 받을 수 없어요!
for _ in range(3):
    links = driver.find_elements(By.TAG_NAME, "a")
    for link in links :
        href = link.get_attribute("href")
        if href and "/p/" in href :
            post_links.add(href)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);") # ; 종료 연산자
    time.sleep(3)

results = []
for url in list(post_links)[:10] : #너무 많은 수는 걸림
    driver.get(url)
    time.sleep(3)
    caption = ""

    try :
        img_tag = driver.find_element(By.XPATH, "//img[@alt]")
        caption = img_tag.get_attribute("alt").strip()
        results.append({"url": url, "caption" : caption})
    except Exception as e :
        print(f"캡션 추출 실패 : {url} | 에러 : {e}")
        continue
        
driver.quit()

for r in results :
    print(r["url"])
    print(r["caption"])
    print("="*60)
```
