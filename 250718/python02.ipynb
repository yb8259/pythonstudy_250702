```python
import requests, openpyxl
# import openpyxl
from bs4 import BeautifulSoup
from openpyxl.styles import Alignment, Font #정렬하다 ~

wrap_alignment = Alignment(wrap_text=True)

header_font = Font(bold=True, size=14)
normal_font = Font(bold=False)

excel_file = openpyxl.Workbook() # 엑셀언더바파일은 인스턴스 객체
excel_sheet = excel_file.active # 시트 생성함
excel_sheet.title = "게시글"

header = ["No","게시글 제목","댓글수"] #하나의 행 안에 여러 값을 넣을땐 리스트 형식으로 한다  
excel_sheet.append(header) # 값을 추가함

for cell in excel_sheet[1] :
    cell.font = header_font

excel_sheet.column_dimensions["A"].width = 20
excel_sheet.column_dimensions["B"].width = 80
excel_sheet.column_dimensions["C"].width = 80


res = requests.get("https://davelee-fun.github.io/trial/board/news.html")
soup = BeautifulSoup(res.content, "html.parser")


#datas = soup.select("div.list_content>div.list_item")
datas = soup.select("div.list_item")
index = 0
row = 2
for data in datas :
    title = data.select_one("span.subject_fixed")
    reply_count = data.select_one("span.rSymph05")
    link = data.select_one("a.list_subject")
    if title != None :
        index +=1
        link_url="https://davelee-fun.github.io/trial/board/"+link["href"]
        # print(index,title.get_text().strip(), [int(reply_count.get_text())])
        data1 = [index,title.get_text().strip(), "["+reply_count.get_text()+"]"]
        excel_sheet.append(data1)
        excel_sheet.cell(row = row, column = 2).hyperlink = link_url # row 는 행, 열
        
        for cell in excel_sheet[row] :# 2번째 행부터 굵게 됨. row가 2니까 반복될때마다 +1 해준것
            cell.font = bold_font 
        row+= 1
        
        res_title = requests.get(link_url)
        soup_title = BeautifulSoup(res_title.content, "html.parser")
        replies = soup_title.select("div.comment_content")
        for reply in replies :
            data2 = ["","",reply.get_text().strip().replace(" ","").replace("\n","").replace("\t","")] # 3열부터 시작함
            # print("  ㄴ",reply.get_text().strip().replace(" ","").replace("\n","").replace("\t","")) 
            excel_sheet.append(data2)
            for cell in excel_sheet[row] :
                cell.font = normal_font 
                cell.alignment = wrap_alignment
            row+=1
            
excel_file.save("mysite.xlsx")
excel_file.close()
print("mysite.xlsx 저장완료!✅")
```

```python
import requests, openpyxl
# import openpyxl
from bs4 import BeautifulSoup
from openpyxl.styles import Alignment, Font, colors #정렬하다 ~

wrap_alignment = Alignment(wrap_text=True) # 정해놓은 셀에 넓ㅅ이가 초과하면 새라인으로 바꿔는 것 정렬!
center_alignment = Alignment(horizontal="center", vertical="center") #가로수평, 세로 수평
bold_font = Font(bold=True)
header_font = Font(bold=True, size=14)
normal_font = Font(bold=False)
hyperlink_font = Font(color=colors.BLUE, underline="single")

excel_file = openpyxl.Workbook() # 엑셀언더바파일은 인스턴스 객체
excel_sheet = excel_file.active # 시트 생성함
excel_sheet.title = "게시글"

header = ["No","게시글 제목","댓글수"] #하나의 행 안에 여러 값을 넣을땐 리스트 형식으로 한다  
excel_sheet.append(header) # 값을 추가함

for cell in excel_sheet[1] :
    cell.font = header_font
    cell.alignment = center_alignment

excel_sheet.column_dimensions["A"].width = 20
excel_sheet.column_dimensions["B"].width = 80
excel_sheet.column_dimensions["C"].width = 80


res = requests.get("https://davelee-fun.github.io/trial/board/news.html")
soup = BeautifulSoup(res.content, "html.parser")


#datas = soup.select("div.list_content>div.list_item")
datas = soup.select("div.list_item")
index = 0
row = 2
for data in datas :
    title = data.select_one("span.subject_fixed")
    reply_count = data.select_one("span.rSymph05")
    link = data.select_one("a.list_subject")
    if title != None :
        index +=1
        link_url="https://davelee-fun.github.io/trial/board/"+link["href"]
        # print(index,title.get_text().strip(), [int(reply_count.get_text())])
        data1 = [index,title.get_text().strip(), "["+reply_count.get_text()+"]"]
        excel_sheet.append(data1)
        excel_sheet.cell(row = row, column = 2).hyperlink = link_url # row 는 행, column은열
        excel_sheet.cell(row = row, column = 2).font = hyperlink_font # 밑에 셀 볼드처리 한다해서 이 코드가 먹지 않음. 그래서 밑에 if절로 상쇄시킴
        excel_sheet.cell(row = row, column = 2).alignment = center_alignment # if절 참고하자!
        
        for cell in excel_sheet[row] :# 2번째 행부터 굵게 됨. row가 2니까 반복될때마다 +1 해준것
            if cell.column != 2 :
                cell.alignment = center_alignment #*
                cell.font = bold_font  #*
        row+= 1
        
        res_title = requests.get(link_url)
        soup_title = BeautifulSoup(res_title.content, "html.parser")
        replies = soup_title.select("div.comment_content")
        for reply in replies :
            data2 = ["","",reply.get_text().strip().replace(" ","").replace("\n","").replace("\t","")] # 3열부터 시작함
            # print("  ㄴ",reply.get_text().strip().replace(" ","").replace("\n","").replace("\t","")) 
            excel_sheet.append(data2)
            for cell in excel_sheet[row] :
                cell.font = normal_font 
                cell.alignment = wrap_alignment
            row+=1
            
excel_file.save("mysite.xlsx")
excel_file.close()
print("mysite.xlsx 저장완료!✅")
```
