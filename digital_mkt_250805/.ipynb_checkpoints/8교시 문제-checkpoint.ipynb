```python
#구글 Scholar에서 'SEO Marketing' 논문 제목을 5개 수집하세요.

import requests
from bs4 import BeautifulSoup
import time

url = "https://scholar.google.com/scholar?hl=ko&as_sdt=0%2C5&q=SEO+marketing&btnG="
response = requests.get(url)
response.encoding = "utf-8"
soup= BeautifulSoup(response.text, "html.parser")
print(soup)
```

```python
#TIOBE Index 사이트에서 상위 10개 프로그래밍 언어 이름을 수집하세요.
import requests
from bs4 import BeautifulSoup
import time

url = "https://www.tiobe.com/tiobe-index/"
response = requests.get(url)
response.encoding = "utf-8"
soup= BeautifulSoup(response.text, "html.parser")

i = 0
indexes = soup.select("table#top20 tr")
for index in indexes : 
    programs = index.find_all("td")
    if len(programs)>=5 :
        program = programs[4].text.strip()
        i += 1
        print(f"{i}. {program}")
        if i == 10 :
            break
```

```python
#데이터마켓(data.go.kr)에서 '소비' 관련 공공데이터 제목을 10개 수집하세요.
import requests
from bs4 import BeautifulSoup
import time

query = "소비"
url = f"https://www.data.go.kr/tcs/dss/selectDataSetList.do?keyword={query}&brm=&svcType=&recmSe=N&conditionType=init&extsn=&kwrdArray="
response = requests.get(url)
time.sleep(3)

response.encoding = "utf-8"
soup= BeautifulSoup(response.text, "html.parser")


elements = soup.select("div.result-list span.title")
for i, element in enumerate(elements) :
    i += 1
    text = element.text.strip()
    print(f"{i}. {text}")
    if i == 10 :
        break
```

```python
#알라딘에서 '자기계발' 베스트 도서의 제목과 저자를 수집하세요.

import requests
from bs4 import BeautifulSoup
import time


url = "https://www.aladin.co.kr/shop/wbrowse.aspx?CID=38400&BrowseTarget=BestSeller"
response = requests.get(url)
time.sleep(3)

response.encoding = "utf-8"
soup= BeautifulSoup(response.text, "html.parser")

i = 0
titles = soup.select("a.bo")
authors = soup.select("span.gw")
for title, author in zip(titles,authors) :
    i += 1
    print(f"{i}. {title.text.strip()} - {author.text.strip()}")
```
